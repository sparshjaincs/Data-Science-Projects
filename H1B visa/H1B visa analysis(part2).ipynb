# Growth Increase per year in Number of Applications for h1b visa analysis

```python
growth=h1['YEAR'].value_counts().sort_values(ascending=True)

growth
```

```python
fig,ax=plt.subplots(figsize=(12,6),facecolor="#123456")
sns.barplot(growth.index,growth.values,ax=ax)
plt.xticks(rotation=90,fontsize=14,color="white")
plt.yticks(fontsize=12,color="white")
plt.title("Growth Increase per year in Number of Applications for h1b visa analysis.",fontsize=16,color="white")
plt.grid()
plt.xlabel("YEARS",fontsize=16,color="white")
plt.ylabel("NUMBER OF APPLICANTS",fontsize=16,color="white")
plt.ylim([0,1000000])
st="YEAR-INTERVAL :INCREMENT(in applicants)\n\n"
for i in range(len(growth.index)-1):
    x1=str(int(growth.index[i]))+" - "+str(int(growth.index[i+1]))
    x2=growth.values[i+1]-growth.values[i]
    st+=f"                {x1} : {x2}\n"
ax.text(0,550000,st,fontsize=14)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,6),facecolor="gray")
sns.barplot(growth.index,growth.values,ax=ax)
plt.xticks(rotation=90,fontsize=8)
plt.yticks(fontsize=12)
plt.title("Growth Increase per year in Number of Applications for h1b visa analysis.",fontsize=16)
plt.grid()
plt.xlabel("YEARS",fontsize=16)
plt.ylabel("NUMBER OF APPLICANTS",fontsize=16)
plt.ylim([0,1000000])
st="YEAR-INTERVAL :INCREMENT(in applicants)\n\n"
for i in range(len(growth.index)-1):
    x1=str(int(growth.index[i]))+" - "+str(int(growth.index[i+1]))
    x2=growth.values[i+1]-growth.values[i]
    st+=f"                {x1} : {x2}\n"
ax.text(0,550000,st,fontsize=14)
plt.savefig("Growth Increase per year in Number of Applications for h1b visa analysis.jpeg")
plt.show()
```

# Deny Rate by Year

```python
rate=h1[['CASE_STATUS',"YEAR"]]
rate=rate[rate['CASE_STATUS']=="DENIED"]
rat=rate['YEAR'].value_counts()

fig,ax=plt.subplots(figsize=(12,6),facecolor="#123456")
sns.barplot(rat.index,rat.values,ax=ax)
plt.xticks(fontsize=14,color="white")
plt.yticks(fontsize=12,color="white")
plt.title("DENY RATE BY YEAR.",fontsize=16,color="white")
plt.grid()
plt.xlabel("YEARS",fontsize=16,color="white")
plt.ylabel("NUMBER OF DENIES",fontsize=16,color="white")
total=int(rate['CASE_STATUS'].value_counts()) # TOTAL NO  OF DENIES
per="YEAR-INTERVAL  : DENY RATE(%))\n\n"
for i in range(len(rat.index)-1):
    
    x1=str(int(rat.index[i]))+" - "+str(int(rat.index[i+1]))
    x2=int(rat.values[i]-rat.values[i+1])
    p=str(round(x2/total,2)*100)+" %"
   
    per+=f"       {x1} : {p}\n"
ax.text(2.5,17111,per,fontsize=14)
    
plt.show()
```

```python
rate=h1[['CASE_STATUS',"YEAR"]]
rate=rate[rate['CASE_STATUS']=="DENIED"]
rat=rate['YEAR'].value_counts()

fig,ax=plt.subplots(figsize=(12,6),facecolor="gray")
sns.barplot(rat.index,rat.values,ax=ax)
plt.xticks(fontsize=14)
plt.yticks(fontsize=12)
plt.title("DENY RATE BY YEAR.",fontsize=16)
plt.grid()
plt.xlabel("YEARS",fontsize=16)
plt.ylabel("NUMBER OF DENIES",fontsize=16)
total=int(rate['CASE_STATUS'].value_counts()) # TOTAL NO  OF DENIES
per="YEAR-INTERVAL  : DENY RATE(%))\n\n"
for i in range(len(rat.index)-1):
    
    x1=str(int(rat.index[i]))+" - "+str(int(rat.index[i+1]))
    x2=int(rat.values[i]-rat.values[i+1])
    p=str(round(x2/total,2)*100)+" %"
   
    per+=f"       {x1} : {p}\n"
ax.text(2.5,17111,per,fontsize=14)
plt.savefig("DENY RATE BY YEAR.jpeg")
plt.show()
```

# What is Average Salary of top 25 Companies ( according to number of applicantes)

```python
h1.head()
```

```python
top1=h1['EMPLOYER_NAME'].value_counts()[:25]
top1
```

```python
df=h1[['EMPLOYER_NAME','PREVAILING_WAGE']]
count=[]
total=[]
for i in top1.index:
    x=df[df['EMPLOYER_NAME']==i]
    count.append(int((x['EMPLOYER_NAME'].value_counts()).values))
    total.append(sum(list(x['PREVAILING_WAGE'])))

    
```

```python
count
```

```python
total
```

```python
m=pd.Series(total)
m[m.isna()]=m.median()
```

```python
list(m)
```

```python
avg=[]
for i,j in zip(count,m):
    avg.append(round(j/i,2))
avg
    
    
```

```python
avg=pd.Series(avg)
```

```python
initials=[]
for i in list(top1.index):
    if len(i)>25:
        s=""
        x=i.split(" ")
        for j in x:
            s+=j[0]
        initials.append(s)
    else:
        initials.append(i)
fig,ax=plt.subplots(figsize=(12,6),facecolor="#123456")
sns.barplot(initials,avg,ax=ax)
plt.xticks(rotation=90,fontsize=14,color="white")
plt.yticks(fontsize=12,color="white")
plt.title("TOP 25 Companies VS Their Average salary.",fontsize=16,color="white")
plt.grid()
plt.xlabel("COMPANIES",fontsize=16,color="white")
plt.ylabel("AVERAGE_SALARY",fontsize=16,color="white")
plt.show()
```

```python
initials=[]
for i in list(top1.index):
    if len(i)>20:
        s=""
        x=i.split(" ")
        for j in x:
            s+=j[0]
        initials.append(s)
    else:
        initials.append(i)
fig,ax=plt.subplots(figsize=(12,6),facecolor="Gray")
sns.barplot(initials,avg,ax=ax)
plt.xticks(rotation=90,fontsize=7)
plt.yticks(fontsize=12)
plt.title("TOP 25 Companies VS Their Average salary.",fontsize=16)
plt.grid()
plt.xlabel("COMPANIES",fontsize=16)
plt.ylabel("AVERAGE_SALARY",fontsize=16)
plt.savefig("TOP 25 Companies VS Their Average salary.jpeg")
plt.show()
```
